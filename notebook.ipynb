## Seção 3 - Métodos

O modelo SIRV(Susceptible-Infectious-Recovered-Vaccinated) seria um modelo adequado pois, além de levar em consideração a população suscetível, infectada e removida - variáveis do modelo SIR -, o modelo também leva em conta a quantidade de pessoas vacinadas. Vamos partir da premissa que o crescimento da população de Pernambuco será desprezível em relação ao total de pessoas, podendo-se assumir portanto que a quantidade de indivíduos só diminui considerando as mortes por COVID e nada a mais. Além disso, vamos considerar também uma função para denotar o número de mortes causadas pela pandemia.

$N=S(t)+I(t)+R(t)+V(t)$

$0= S'+I'+R'+V'$

$\frac{dS(t)}{dt} =-\frac{β}{N}S(t)I(t)-αS(t)$

$\frac{dI(t)}{dt} =\frac{β}{N}S(t)I(t)-(γ+μ)I(t)$

$\frac{dR(t)}{dt} =γI(t) - αR(t)$

$\frac{dV(t)}{dt} =αS(t) + αR(t)$

$\frac{dD(t)}{dt}= μI(t)$



Em que:

-**N é a população total**
N = 9.278 milhões de pessoas (Senso - IBGE)

-**β é a taxa de transmissão diária da doença para os não vacinados**
β = 0,89 , de acordo com [2]

-**γ é a taxa de recuperados diária. 1/γ nos informa a média de dias que uma pessoa fica doente**
1/γ = 14 dias, daí γ = 1/14

-**α é a taxa de vacinação, levando em consideração a eficácia das vacinas**
Taxa de eficácia = 0,765 (Pfizer = 95%, Astrazeneca = 82,4%, CoronaVac = 50,38%, Jansen = 85%. α = (0.95*4.9 + 0.824*4.7 + 0.5038*4.5 + 0.85*0.17)/(14.3))

Não temos dados suficientes para calcular precisamente a taxa de vacinação de pernambuco, porém temos os dados para calcular a taxa de vacinação de Recife. Vamos usar uma aproximação que a taxa de vacinação média do estado de Pernambuco é 30% menor que a taxa de vacinação de recife.

Taxa de vacinação de recife:
Nos últimos dias, aproximadamente 9506 pessoas estavam sendo vacinadas por dia em Recife, de um total de aproximadamente 1.661.017 habitantes. Daí, a temos uma taxa de 9506/1661017 = 0,0057 para Recife

Daí, considerando que a taxa de vacinação de Pernambuco é 30% menor que a de Recife, essa primeira seria dada por aproximadamente 0,004

α = 0,004*0,765


-**μ é a taxa de mortalidade diária da Covid-19**

O cálculo foi feito pegando o valor de mortes na última semana em [5], dividindo o valor por 7 e, por fim, dividindo ele pela população brasileira(212 milhões). Aqui foi feita uma aproximação do valor do brasil para o de pernambuco.

-**S(t) é a população ainda suscetível a pegar a doença no tempo t (não vacinados)**
Para calcular a quantidade de susceptíveis do dia inicial (13/11/2021) podemos pegar a população total desse dia e tirar a quantidade de infectados, recuperados, vacinados e mortos

-**I(t) é o total da população infectada no o tempo t**
Podemos obter o número de pessoas infectadas atualmente somando o número de novos casos há 14 dias. Portanto, considerando os novos casos desde o dia 30 de outubro, resultando em 4508 pessoas infectadas atualmente.

-**R(t) é a população removida até o tempo t. Ou seja, a população que se recuperou da doença**
No dia 13/11/2021 Pernambuco tinha um total de 68.496 recuperados

-**V(t) é o total da população vacinada com as duas doses até o tempo t**
No dia 13/11/2021 Pernambuco tinha um total de 5.009.762 de vacinados com duas doses


-**D(t) é o total de óbitos em decorrência da Covid-19**
Como vamos contar as mortes a partir desse dia, vamos começar a contar a partir de 0 mortos




```
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt
# Setando as constantes

beta = 0.89
gamma = 0.07143
alpha = 0.004*0.765
mi = 0.00000147
N = 9278000

I0 = 4508
R0 = 68496
V0 = 5009762
D0 = 0
S0 = N - I0 - R0 - V0


t = np.linspace(0, 120, 120) # Vamos ver qual será a progressão da doença nos próximos 4 meses
```

```
# Código de métodos

def modelo_SIR(y, t):
    S, I, R, D = y
    dSdt = (-beta * S * I / N)
    dIdt = (beta * S * I / N) - (gamma * I) - mi * I
    dRdt = (gamma * I)
    dDdt = mi * I
    return dSdt, dIdt, dRdt, dDdt

def modelo_SIRVD(y, t):
    S, I, R, V, D = y
    dSdt = (-beta * S * I / N) - S*alpha
    dIdt = (beta * S * I / N) - (gamma * I) - mi * I
    dRdt = (gamma * I) - alpha*R
    dVdt = alpha * (S + R)
    dDdt = mi * I
    return dSdt, dIdt, dRdt, dVdt, dDdt
```

### Prevendo a evolução dos casos nas situações atuais

Considerando a vacinação e a taxa de transmissão atual, assim seria dada a evolução dos casos de COVID em Pernambuco nos próximos 120 dias (a partir de 13/11/2021)

```
y0 = (S0, I0, R0, V0, D0)

# Resolvendo as equações numericamente e obtendo os resultados do modelo
results = odeint(modelo_SIRVD, y0, t)
S, I, R, V, D = results.T

plt.suptitle("Modelo SIRV")
plt.plot(t, S/N, 'g', label = 'Susceptiveis')
plt.plot(t, I/N, 'y', label = 'Infectados')
plt.plot(t, R/N, 'b', label = 'Recuperados')
plt.plot(t, V/N, 'c', label = 'Vacinados')
plt.plot(t, D/N, 'r', label = 'Mortos')
plt.xlabel('Dias a partir de 13/11/2021')
plt.ylabel('Fração da população')
plt.grid(b=True, which='major', c='#eeeeee', lw=2, ls='-')
plt.legend(['Susceptiveis','Infectados', 'Recuperados', 'Vacinados', 'Mortos'], loc='upper right')

plt.show()
```

#### Mortes

```
plt.suptitle("Mortes em número absoluto")
plt.plot(t, D, 'r', label = 'Mortos')
plt.xlabel('Dias a partir de 13/11/2021')
plt.ylabel('Número absoluto de mortes')
plt.grid(b=True, which='major', c='#eeeeee', lw=2, ls='-')
plt.legend(['Mortos'], loc='upper right')

plt.show()
```

### Prevendo a evolução dos casos sem considerar as vacinações

Aqui está ilustrada uma comparação entre o número de infectados e o número de mortes caso desconsiderássemos o efeito da vacina nos pernambucanos já vacinados. Para isso, utilizamos o modelo SIR, considerando o número de mortos, e considerando que os vacinados são susceptíveis, desconsiderando, dessa forma, o efeito da vacina.

```
y0_SIR = (S0 + V0, I0, R0, D0)

# Resolvendo as equações numericamente e obtendo os resultados do modelo
results_SIR = odeint(modelo_SIR, y0_SIR, t)
S_SIR, I_SIR, R_SIR, D_SIR = results_SIR.T

fig, axs = plt.subplots(1, 2, figsize=(12, 5), constrained_layout=True,
                        sharex=True, sharey=False, facecolor='#eeeeee')

axs.flat[0].plot(t, I/N, 'y', label = 'Considerando o efeito da vacina')
axs.flat[0].plot(t, I_SIR/N, 'm', label = 'Desconsiderando o efeito da vacina')
axs.flat[0].set_xlabel('Dias a partir de 13/11/2021')
axs.flat[0].set_ylabel('Fração da população')
axs.flat[0].grid(b=True, which='major', c='#eeeeee', lw=2, ls='-')
axs.flat[0].legend(['Considerando o efeito da vacina','Desconsiderando o efeito da vacina'], loc='upper right')

axs.flat[1].plot(t, D, 'y', label = 'Considerando o efeito da vacina')
axs.flat[1].plot(t, D_SIR, 'm', label = 'Desconsiderando o efeito da vacina')
axs.flat[1].set_xlabel('Dias a partir de 13/11/2021')
axs.flat[1].set_ylabel('Fração da população')
axs.flat[1].grid(b=True, which='major', c='#eeeeee', lw=2, ls='-')
axs.flat[1].legend(['Considerando o efeito da vacina','Desconsiderando o efeito da vacina'], loc='upper right')

plt.show()
```

### Prevendo a evolução dos casos considerando um aumento na taxa de transmissão

A seguir, fazemos a previsão do número de infectados considerando a vacinação, porém considerando agora uma taxa de transmissão elevada (Beta = 1,30, na qual denota que 100 infectados transmitem a doença em média para 130 outras pessoas). Essa foi uma das taxas de transmissão mais altas que já tivemos no Brasil (ver na referência 6). É interessante analisarmos o impacto disso, visto que o governo anda reduzindo as medidas restritivas e estamos nos aproximando das festas de fim de ano, o que certamente provocará um aumento na taxa de transmissão do vírus, em relação a taxa atual. 

```
y0 = (S0, I0, R0, V0, D0)

beta = 1.3 # Taxa máxima de transmissão já registrada no Brasil

# Resolvendo as equações numericamente e obtendo os resultados do modelo
results_mod = odeint(modelo_SIRVD, y0, t)
S_mod, I_mod, R_mod, V_mod, D_mod = results_mod.T

plt.suptitle("Modelo SIRV comparação de betas")
plt.plot(t, I, 'y', label = 'Taxa de transmissão atual')
plt.plot(t, I_mod, 'm', label = 'Taxa de transmissão aumentada')
plt.xlabel('Dias a partir de 13/11/2021')
plt.ylabel('Número absoluto de infectados')
plt.grid(b=True, which='major', c='#eeeeee', lw=2, ls='-')
plt.legend(['Taxa de transmissão atual','Taxa de transmissão aumentada'], loc='upper right')
```

## Seção 4 - Resultados e Discussão

A partir da execução e cálculo do modelo SIRV sob as circunstâncias do estado de Pernambuco foi possível chegar à previsão que a pandemia pode chegar a um fim em aproximadamente 100 dias. Além disso, é notável que as mortes vão chegar em um teto perto de 80  no estado. Outro dado interessante é que a população de vacinados pode chegar até 70% em 120 dias, um ganho de 15% aproximadamente. Ademais, os valores que foram utilizados para calibrar o nosso modelo se referem a uma janela do tempo, o dia de hoje, portanto, eles devem mudar e não representar a realidade tão bem.

É claro que essa previsão pode desviar da realidade, pois estamos chegando em uma época de fim de ano, em que há uma grande tendência para um aumento de casos durante as festividades. Além disso, a diminuição do uso de máscara também pode incrementar esse número de infectados e mortos. Apesar disso, a relevância da expansão da campanha de vacinação é demonstrada tanto na nossa previsão matemática como na vida real. Com o crescimento do número de vacinados com a segunda dose acima de 70%, espera-se que os casos de covid cheguem a números próximos de 0%. Mesmo com um modelo que não represente a realidade com completa fidelidade, pode-se observar essa tendência que acreditamos condizer com o futuro próximo. Uma boa referência para entender o impacto da vacinação é o nosso gráfico feito com o modelo SIR, considerando que ninguém foi vacinado e que ninguem será, nesse cenário observa-se que a quantidade de infectados chega próxima de 70% e que o número de mortes chega próximo de 190.

Para tentar simular um pouco como seria o cenário de fim de ano  e de menor relevância dos costumes anti-covid, utilizamos uma outra versão do nosso modelo SIRV com a taxa de transmissão, Beta, modificada para 1,3. Ao criar esse cenário o número de infectados aumentou em aproximadamente 25%, apesar de não ser precisa, essa estimativa pode nos preparar para o que pode vir no futuro próximo.
