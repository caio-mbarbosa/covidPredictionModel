## Seção 3 - Métodos

O modelo SIRV(Susceptible-Infectious-Recovered-Vaccinated) seria um modelo adequado pois, além de levar em consideração a população suscetível, infectada e removida - variáveis do modelo SIR -, o modelo também leva em conta a quantidade de pessoas vacinadas. Vamos partir da premissa que o crescimento da população de Pernambuco será desprezível em relação ao total de pessoas, podendo-se assumir portanto que a quantidade de indivíduos só diminui considerando as mortes por COVID e nada a mais. Além disso, vamos considerar também uma função para denotar o número de mortes causadas pela pandemia.

$N=S(t)+I(t)+R(t)+V(t)$

$0= S'+I'+R'+V'$

$\frac{dS(t)}{dt} =-\frac{β}{N}S(t)I(t)-αS(t)$

$\frac{dI(t)}{dt} =\frac{β}{N}S(t)I(t)-(γ+μ)I(t)$

$\frac{dR(t)}{dt} =γI(t) - αR(t)$

$\frac{dV(t)}{dt} =αS(t) + αR(t)$

$\frac{dD(t)}{dt}= μI(t)$



Em que:

-**N é a população total**
N = 9.278 milhões de pessoas (Senso - IBGE)

-**β é a taxa de transmissão diária da doença para os não vacinados**
β = 0,89 , de acordo com [2]

-**γ é a taxa de recuperados diária. 1/γ nos informa a média de dias que uma pessoa fica doente**
1/γ = 14 dias, daí γ = 1/14

-**α é a taxa de vacinação, levando em consideração a eficácia das vacinas**
Taxa de eficácia = 0,765 (Pfizer = 95%, Astrazeneca = 82,4%, CoronaVac = 50,38%, Jansen = 85%. α = (0.95*4.9 + 0.824*4.7 + 0.5038*4.5 + 0.85*0.17)/(14.3))

Não temos dados suficientes para calcular precisamente a taxa de vacinação de pernambuco, porém temos os dados para calcular a taxa de vacinação de Recife. Vamos usar uma aproximação que a taxa de vacinação média do estado de Pernambuco é 30% menor que a taxa de vacinação de recife.

Taxa de vacinação de recife:
Nos últimos dias, aproximadamente 9506 pessoas estavam sendo vacinadas por dia em Recife, de um total de aproximadamente 1.661.017 habitantes. Daí, a temos uma taxa de 9506/1661017 = 0,0057 para Recife

Daí, considerando que a taxa de vacinação de Pernambuco é 30% menor que a de Recife, essa primeira seria dada por aproximadamente 0,004

α = 0,004*0,765


-**μ é a taxa de mortalidade diária da Covid-19**

O cálculo foi feito pegando o valor de mortes na última semana em [5], dividindo o valor por 7 e, por fim, dividindo ele pela população brasileira(212 milhões). Aqui foi feita uma aproximação do valor do brasil para o de pernambuco.

-**S(t) é a população ainda suscetível a pegar a doença no tempo t (não vacinados)**
Para calcular a quantidade de susceptíveis do dia inicial (13/11/2021) podemos pegar a população total desse dia e tirar a quantidade de infectados, recuperados, vacinados e mortos

-**I(t) é o total da população infectada no o tempo t**
Podemos obter o número de pessoas infectadas atualmente somando o número de novos casos há 14 dias. Portanto, considerando os novos casos desde o dia 30 de outubro, resultando em 4508 pessoas infectadas atualmente.

-**R(t) é a população removida até o tempo t. Ou seja, a população que se recuperou da doença**
No dia 13/11/2021 Pernambuco tinha um total de 68.496 recuperados

-**V(t) é o total da população vacinada com as duas doses até o tempo t**
No dia 13/11/2021 Pernambuco tinha uma cobertura completa de 79,5%, de acordo com [8]


-**D(t) é o total de óbitos em decorrência da Covid-19**
Como vamos contar as mortes a partir desse dia, vamos começar a contar a partir de 0 mortos




```
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt
# Setando as constantes

eficacia_media_vacinas = 0.765

beta = 0.89
gamma = 0.07143
alpha = 0.004*eficacia_media_vacinas
mi = 0.0000147
N = 9278000

I0 = 4508
R0 = 68496
numero_absoluto_vacinados = N*0.7951                      #NOTE que o numero absoluto de vacinados não é o nosso V0! V0 é o numero de vacinados de forma EFICAZ! Precisamos considerar a eficacia de cada vacina
V0 = numero_absoluto_vacinados*eficacia_media_vacinas     #V0 considera a eficacia de cada vacina!
D0 = 0
S0 = N - I0 - R0 - V0


t = np.linspace(0, 120, 120) # Vamos ver qual será a progressão da doença nos próximos 4 meses
```

```
# Código de métodos

def modelo_SIR(y, t):
    S, I, R, D = y
    dSdt = (-beta * S * I / N)
    dIdt = (beta * S * I / N) - (gamma * I) - mi * I
    dRdt = (gamma * I)
    dDdt = mi * I
    return dSdt, dIdt, dRdt, dDdt

def modelo_SIRVD(y, t):
    S, I, R, V, D = y
    dSdt = (-beta * S * I / N) - S*alpha
    dIdt = (beta * S * I / N) - (gamma * I) - mi * I
    dRdt = (gamma * I) - alpha*R
    dVdt = alpha * (S + R)
    dDdt = mi * I
    return dSdt, dIdt, dRdt, dVdt, dDdt

def modelo_SIRVD_omicron(y, t):
    S, I, R, V, D = y
    dSdt = (-(beta_omicron(t)) * S * I / N) - S*alpha
    dIdt = (beta_omicron(t) * S * I / N) - (gamma * I) - mi * I
    dRdt = (gamma * I) - alpha*R
    dVdt = alpha * (S + R)
    dDdt = mi * I
    return dSdt, dIdt, dRdt, dVdt, dDdt

```

### Prevendo a evolução dos casos nas situações atuais

Considerando a vacinação e a taxa de transmissão atual, assim seria dada a evolução dos casos de COVID em Pernambuco nos próximos 120 dias (a partir de 13/11/2021)

```
y0 = (S0, I0, R0, V0, D0)

# Resolvendo as equações numericamente e obtendo os resultados do modelo
results = odeint(modelo_SIRVD, y0, t)
S, I, R, V, D = results.T

fig, axs = plt.subplots(1, 2, figsize=(12, 5), constrained_layout=True,
                        sharex=True, sharey=False, facecolor='#eeeeee')

axs.flat[0].plot(t, S/N, 'g', label = 'Susceptiveis')
axs.flat[0].plot(t, I/N, 'y', label = 'Infectados')
axs.flat[0].plot(t, R/N, 'b', label = 'Recuperados')
axs.flat[0].plot(t, V/N, 'c', label = 'Vacinados')
axs.flat[0].plot(t, D/N, 'r', label = 'Mortos')
axs.flat[0].set_xlabel('Dias a partir de 13/11/2021')
axs.flat[0].set_ylabel('Fração da população')
axs.flat[0].grid(b=True, which='major', c='#eeeeee', lw=2, ls='-')
axs.flat[0].legend(['Susceptiveis','Infectados', 'Recuperados', 'Vacinados', 'Mortos'], loc='upper right')


axs.flat[1].plot(t, D, 'r', label = 'Mortos')
axs.flat[1].set_xlabel('Dias a partir de 13/11/2021')
axs.flat[1].set_ylabel('Número acumulado de mortes')
axs.flat[1].grid(b=True, which='major', c='#eeeeee', lw=2, ls='-')
axs.flat[1].legend(['Mortos'], loc='upper right')

plt.show()

# Pegando a maior fração da população infectada em um dia

print("O maior número de pessoas na classe de 'infectados' em um dia é: "+ str(round(max(I)/1000000, 2)) + " milhões de pessoas")

# Pegando a quantidade de mortos depois de 120 dias
print("O número de mortos depois de 120 dias é de: " + str(int(max(D))) + " mortos")

```
